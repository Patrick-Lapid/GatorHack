```python
import requests
import xml.etree.ElementTree as ET
import csv

# List of XML sitemap URLs
sitemap_urls = [
    "https://www.verizon.com/sitemapls.xml",
    "https://www.verizon.com/sitemap-mobile.xml",
    "https://www.verizon.com/sitemap-products.xml",
    "https://www.verizon.com/sitemap-support-home.xml",
    "https://www.verizon.com/sitemap-support-mobile.xml",
    "https://www.verizon.com/sitemap-kb.xml",
    "https://www.verizon.com/sitemap-kb2.xml",
    "https://www.verizon.com/sitemap-kb3.xml",
    "https://www.verizon.com/sitemap-stores-old.xml",
    "https://www.verizon.com/about/sitemap.xml",
    "https://www.verizon.com/business/sitemap.xml",
    # Add more sitemap URLs as needed
]


unique_urls = set()

# Loop through the list of sitemap URLs
for sitemap_url in sitemap_urls:
    # Send a GET request to the sitemap URL
    response = requests.get(sitemap_url)

    if response.status_code == 200:
        # Parse the XML content
        xml_content = response.text
        root = ET.fromstring(xml_content)

        # Find all 'url' elements in the sitemap
        url_elements = root.findall(".//{http://www.sitemaps.org/schemas/sitemap/0.9}url")

        # Extract and add the URLs to the list
        for url_element in url_elements:
            loc_element = url_element.find("{http://www.sitemaps.org/schemas/sitemap/0.9}loc")
            if loc_element is not None and loc_element.text.find("espanol") == -1:
                url = loc_element.text
                unique_urls.add(url)
    else:
        print(f"Failed to retrieve sitemap {sitemap_url}. Status code:", response.status_code)

# Save all the extracted URLs to a CSV file
with open('sitemap_urls.csv', 'w', newline='') as csvfile:
    fieldnames = ['URL']
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)

    # Write the header row
    writer.writeheader()

    for url in unique_urls:
        writer.writerow({'URL': url})
```

```python
import csv
import os

with open('urls1.csv', 'r') as file:
    reader = csv.reader(file, delimiter=',')
    next(reader)  # skip the header row
    data = list(reader)

all_data = []
for idx, (url, name) in enumerate(data):
    # Step 2: For each name, try to read the corresponding text file
    try:
        with open(os.path.join('text', f"{name}.txt"), 'r', encoding="utf-8") as txt_file:
            raw_text = txt_file.read().replace("\n", " ").replace("\r", " ").strip()  # replacing newlines with spaces
            # Ensure there's no newline at the end
            raw_text = raw_text.replace(",", " ")
            raw_text = raw_text.rstrip("\n").rstrip("\r")
            
    except FileNotFoundError:
        print(f"Warning: File for {name} not found.")
        raw_text = ''  # or continue if you want to skip rows without corresponding files

    all_data.append((idx, url, raw_text))

with open('all_data.csv', 'w', newline='', encoding="utf-8") as csvfile:
    writer = csv.writer(csvfile, delimiter=',')
    writer.writerow(['id', 'url', 'raw text'])  # write the header
    for row in all_data:
        writer.writerow(row)
```
